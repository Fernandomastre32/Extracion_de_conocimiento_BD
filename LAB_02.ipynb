```python
import pandas as pd
```

```python
import random
```

```python
cantidad = 1000000
```

```python
nombres = ["Ana", "Luis", "Martha", "Carolina", "Sofía", "Pedro", "Elena", "Diego"]
tipos_post = ["Imagen", "Video", "Reel", "Carrusel", "Story"]
```

```python
df_social = pd.DataFrame({
    "influencer": [random.choice(nombres) for _ in range(cantidad)],
    "likes": [random.randint(100, 50000) for _ in range(cantidad)],
    "comentarios": [random.randint(10, 5000) for _ in range(cantidad)],
    "anio_post": [random.randint(2020, 2026) for _ in range(cantidad)],
    "tipo_post": [random.choice(tipos_post) for _ in range(cantidad)]
})
```

```python
print(df_social.head())
```

```python
df_social.memory_usage(deep = True)
```
