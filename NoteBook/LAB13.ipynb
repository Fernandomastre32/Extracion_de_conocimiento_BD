```python
# Importar librerías de manipulación de datos y visualización
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Importar kagglehub para descargar el dataset
import kagglehub
```

```python
from sklearn.metrics import (
accuracy_score,
confusion_matrix,
classification_report, ConfusionMatrixDisplay
)
```

```python
# Cargamos el data set iris
from sklearn.datasets import load_iris
iris = load_iris(as_frame = True)
```

```python
## Renombrar la columna target
dfiris.rename(columns = {"target" : "Especie"}, inplace = True)
```

```python
dfiris["Especie"] = dfiris["Especie"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})
```

```python
dfiris.head()
```

```python
dfiris.tail()
```

## Identificar variables predictoras y variables de objetivo

```python
# Variable predictoria

X = dfiris.drop(columns = ["Especie"])
```

```python
print("Variables predictorias")
print(X.columns)
```

```python
# Variable objetivo
y = dfiris["Especie"]
```

```python
print("Variables Objetivo")
print(y.name)
```

## Dividir los datos en prueba o entrenamiento

80% para entrenamiento 

20% para prueba

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

## Escalado de variables

```python
scaler = StandardScaler()
```

```python
#Escalamos las variables
X_train_scaler = scaler.fit_transform(X_train)
```

```python
X_test_scaler = scaler.transform(X_test)
```

```python
print("Sin escalar")
print(X_train.head())
print("Escalado")
print(X_train_scaler[:5])
```

## Entrenamiento del modelo KNN
En esta prueba utilizaremos k = 3

```python
modelo = KNeighborsClassifier(n_neighbors=3)
```

```python
modelo.fit(
    X_train_scaler,
    y_train
)
```

## Realizando predicciones con las variables de prueba owo


```python
prediccion = modelo.predict(X_test_scaler)
```

```python
print(prediccion[:10])
```

```python
#Comparacion entre valores reales y valores predecidos

Comparacion = pd.DataFrame({
    "Especie real " : y_test.values,
    "Especie predicha" : prediccion
})
```

```python
print(Comparacion.head(10))
```

```python
print(Comparacion.head(30))
```

## Evaluacion del modelo

```python
accuracy = accuracy_score(y_test, prediccion)
print(f"accuracyExactitud: {accuracy: 47}")
```

```python
##Matriz de confusion 
cm = confusion_matrix(y_test, prediccion)
```

```python
cm_df= pd.DataFrame(
    cm, 
    index = modelo.classes_ ,
    columns = modelo.classes_
)
cm_df
```

## Clasification report
Proporciona metricas que permite evaluar el desempeño de cada clase : precision, recall, f1-scord, support

```python
print(classification_report( y_test, prediccion))
```
